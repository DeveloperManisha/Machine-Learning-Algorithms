```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
X= dataset.iloc[:,3:5].values
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters= 5, init='k-means++', random_state=0)
Y_out=model.fit_predict(X)
```

```python
#visualizing clusters
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.scatter(X[Y_out == 0, 0], X[Y_out == 0, 1], s = 100, c = 'red', label = 'High earning,less expenditure')
plt.scatter(X[Y_out == 1, 0], X[Y_out == 1, 1], s = 100, c = 'blue', label = 'avg earning,avg expenditure')
plt.scatter(X[Y_out == 2, 0], X[Y_out == 2, 1], s = 100, c = 'green', label = 'high earning,high expenditure')
plt.scatter(X[Y_out == 3, 0], X[Y_out == 3, 1], s = 100, c = 'cyan', label = 'less earning,high expenditure')
plt.scatter(X[Y_out == 4, 0], X[Y_out == 4, 1], s = 100, c = 'magenta', label = 'less earning,less expenditure')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Customer Segmentation')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
